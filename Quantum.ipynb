```python
import numpy as np
import matplotlib.pyplot as plt

# Define the states and probabilities
states = ['0', '1']
bit_states = [1, 0]             # Deterministic classical bit
qubit_probs = [0.5, 0.5]        # Equal superposition for qubit

# Create side-by-side bar charts
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

axes[0].bar(states, bit_states)
axes[0].set_title('Classical Bit')
axes[0].set_ylabel('Probability')
axes[0].set_ylim(0, 1)

axes[1].bar(states, qubit_probs)
axes[1].set_title('Qubit in Superposition')
axes[1].set_ylim(0, 1)

plt.tight_layout()
plt.show()

# Simulate measurements of the qubit
num_shots = 1000
results = np.random.choice(states, size=num_shots, p=qubit_probs)
counts = {state: np.count_nonzero(results == state) for state in states}

# Display the simulation results
print(f"Measurement results over {num_shots} shots:")
for state, count in counts.items():
    print(f"State {state}: {count} times (≈ {count/num_shots:.2f} probability)")
```
